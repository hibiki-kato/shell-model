```python
import os
import re
import numpy as np
from matplotlib import pyplot as plt

folder_path = '../laminar_initials'
parameters=[]

for file_name in os.listdir(folder_path):
    beta_nu = re.findall(r'beta([\d\.]+)_nu([\d\.]+)_', file_name)
    if beta_nu:
        beta, nu = map(float, beta_nu[0])
        parameters.append([beta, nu])
parameters = np.array(parameters)
```

```python

fig, ax = plt.subplots()
ax.set_title(r"Relationship between $\nu$ and $\beta$")
ax.set_xlabel("beta")
ax.set_ylabel("nu")
ax.set_xticks(np.linspace(0.42,0.50,33))
ax.xaxis.set_tick_params(rotation=90)
ax.set_yticks(np.linspace(0.00011,0.00018,14))
ax.scatter(parameters[:,0], parameters[:,1],s=1)
ax.grid(True)
fig.savefig("Relationship_between_nu_and_beta")
```

```python
# "/Users/hibiki/Documents/cpp/sync/"
import os
import re
import numpy as np
from matplotlib import pyplot as plt

folder_path = '/Users/hibiki/Documents/cpp/sync'
parameters=[]
for file_name in os.listdir (folder_path):
    # beta_0.4nu_0.000201959_100000period1500window99000sync.npy
    # beta, nu, sync
    beta_nu_period = re.findall(r'beta_([\d\.]+(?:[eE][-+]?\d+)?)nu_([\d\.]+(?:[eE][-+]?\d+)?).*window([\d\.]+(?:[eE][-+]?\d+)?)sync', file_name)
    if beta_nu_period:
        beta, nu, period = map(float, beta_nu_period[0])
        parameters.append([beta, nu, period+1])
parameters = np.array(parameters)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("beta")
ax.set_ylabel("nu")
ax.set_yscale('log')
# parameters[:, 2]
scatter = ax.scatter(parameters[:,0], parameters[:,1], s=10, c=np.log10(parameters[:,2]))
# 
plt.colorbar(scatter)
```
